```python
# this section should be repeated for each input
prev_output_nonce = wally.tx_get_output_nonce(input_tx, 0)
prev_output_rangeproof = wally.tx_get_output_rangeproof(input_tx, 0)
prev_output_script = wally.tx_get_output_script(input_tx, 0) # "extra"
prev_output_asset_tag = wally.tx_get_output_asset(input_tx, 0)
prev_output_commitment = wally.tx_get_output_value(input_tx, 0)

print("sk: {}".format(wally.hex_from_bytes(our_blinding_key)))

satoshi, asset_id, abf, vbf = wally.asset_unblind(prev_output_nonce, our_blinding_key, prev_output_rangeproof, prev_output_commitment, prev_output_script, prev_output_asset_tag)

input_assets = [asset_id]
input_abfs = [abf]
input_ags = [wally.asset_generator_from_bytes(asset_id, abf)]
input_vbfs = [vbf]
input_values = [satoshi, satoshi_out_amount] # both input and output values???

num_outputs = 1

output_abfs = [bytearray(os.urandom(32)) for _ in range(num_outputs)]
output_vbfs = [bytearray(os.urandom(32)) for _ in range(num_outputs - 1)]

_abfs = functools.reduce(lambda x,y: x+y, input_abfs, b'')
_vbfs = functools.reduce(lambda x,y: x+y, input_vbfs, b'')

# copy stuff from the input and then push the outputs at the end
_abfs_with_out = _abfs + functools.reduce(lambda x,y: x+y, output_abfs, b'')
_vbfs_with_out = _vbfs + functools.reduce(lambda x,y: x+y, output_vbfs, b'')

_assets = functools.reduce(lambda x,y: x+y, input_assets)
_ags = functools.reduce(lambda x,y: x+y, input_ags)

output_vbfs += [wally.asset_final_vbf(input_values, 1, _abfs_with_out, _vbfs_with_out)]

# this section should be repeated for each output
eph_sk = os.urandom(32)
eph_pk = wally.ec_public_key_from_private_key(eph_sk)

generator = wally.asset_generator_from_bytes(BITCOIN, output_abfs[0])
value_commitment = wally.asset_value_commitment(satoshi_out_amount, output_vbfs[0], generator)
rangeproof = wally.asset_rangeproof(
    satoshi_out_amount,
    to_blind_key,
    eph_sk,
    BITCOIN,
    output_abfs[0],
    output_vbfs[0],
    value_commitment,
    to_scriptPubKey,
    generator,
    1, # min_value
    0, # exponent
    36 # bits
)

surjectionproof = wally.asset_surjectionproof(
    BITCOIN,
    output_abfs[0],
    generator,
    os.urandom(32),
    _assets,
    _abfs,
    _ags
)

# add our blinded output
wally.tx_add_elements_raw_output(
    tx,
    to_scriptPubKey,
    generator,
    value_commitment,
    eph_pk, # nonce
    surjectionproof,
    rangeproof,
    0
)

# and the explicit fee output
wally.tx_add_elements_raw_output(
    tx,
    None,
    bytearray([0x01]) + BITCOIN,
    wally.tx_confidential_value_from_satoshi(80000000 - satoshi_out_amount),
    None, # nonce
    None, # surjection proof
    None, # range proof
    0)
```
